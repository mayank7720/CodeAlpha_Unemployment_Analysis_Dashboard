# Exploratory Data Analysis (EDA) - Unemployment in India

This notebook contains the exploratory data analysis for the unemployment datasets in India during the period of 2019-2020. The primary goal is to understand the distribution, regional differences, and the impact of the COVID-19 pandemic on unemployment.

## 1. Imports and Data Ingestion

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df_state = pd.read_csv('../data/raw/Unemployment_Rate_upto_11_2020.csv')
df_area = pd.read_csv('../data/raw/Unemployment_in_India.csv')

print("State dataset shape:", df_state.shape)
print("Area dataset shape:", df_area.shape)
```

## 2. Data Ingestion & Cleaning
- Clean column names (strip whitespace)
- Handle missing/null values
- Format datetimes

```python
df_state.columns = df_state.columns.str.strip()
df_area.columns = df_area.columns.str.strip()

df_area.dropna(inplace=True)

df_state['Date'] = pd.to_datetime(df_state['Date'].str.strip(), format='%d-%m-%Y')
df_area['Date'] = pd.to_datetime(df_area['Date'].str.strip(), format='%d-%m-%Y')

print("Cleaned Datetime Ranges:")
print("State data dates:", df_state['Date'].min(), "to", df_state['Date'].max())
print("Area data dates:", df_area['Date'].min(), "to", df_area['Date'].max())
```

## 3. General Distributions & Summaries

```python
print(df_state.describe())
print(df_area.describe())
```

## 4. COVID-19 Lockdown Impact Analysis
Let's check the pre-lockdown average vs. the peak lockdown average (April-May 2020).

```python
lockdown_mask = (df_state['Date'] >= '2020-03-01') & (df_state['Date'] <= '2020-05-31')
pre_lockdown_mask = (df_state['Date'] < '2020-03-01')

print("Pre-COVID average rate:", df_state[pre_lockdown_mask]['Estimated Unemployment Rate (%)'].mean())
print("Lockdown peak average rate:", df_state[lockdown_mask]['Estimated Unemployment Rate (%)'].mean())
```
